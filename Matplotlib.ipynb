```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

_Note: If you're not using Jupyter Notebook, use `plt.show()` at the end of the plot to show the plot._

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

## Functional Plot

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)     # (row, col, plot num)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'g')
```

## Object Oriented

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

canvas = fig.add_axes([0, 0, 1, 1])            # Canvas usually doesn't have axes & stays blank.
canvas.set_title('Canvas')

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.2, 0.8, 0.8])     # (Canvas X, Canvas Y, Length of X-axis, Length of Y-axis)

axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

n = 1
for curr_axes in axes:
    curr_axes.plot(x,y)
    curr_axes.set_title('Plot ' + str(n))
    n+=1

plt.tight_layout()
```

### Figure Size & DPI

```python
fig = plt.figure(figsize=(4,2), dpi=120)

ax = fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')
ax.legend(loc=0)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Saving Plot

```python
fig.savefig('my_pic.jpg', dpi=120)
```

### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=5,alpha=0.5,linestyle='--',marker='o',markersize=5)
#ax.plot(x,y,color='red',lw=5,alpha=0.5,ls='--',marker='o',markersize=5)

ax.set_xlim([0,5])
ax.set_ylim([0,5])
```
